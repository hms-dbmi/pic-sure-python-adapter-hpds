```python
import picsure
```

```python
token_file = "token.txt"

with open(token_file, "r") as f:
    my_token = f.read()
```

```python
authorized_session = picsure.connect(
    picsure.Platform.NHANES_AUTHORIZED,
    my_token
)
```

```python
authorized_session.searchDictionary("age")
```

```python
authorized_session.showAllFacets()
```

```python
child_facet = authorized_session.facets()
child_facet.add("Consortium_Curated_Facets", '06m-post index')
child_facet.view()

authorized_session.searchDictionary("", facets=child_facet)
```

```python
facets = authorized_session.facets()
facets.add("dataset_id", "phs000007")
facets.view()
```

```python
authorized_session.searchDictionary("sex", facets=facets)
```
