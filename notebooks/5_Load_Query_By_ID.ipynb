```python
import picsure
```

```python
token_file = "token.txt"

with open(token_file, "r") as f:
    my_token = f.read()

session = picsure.connect(
    picsure.Platform.BDC_AUTHORIZED,
    dev_mode=True,
    token=my_token
)
```

```python
# loadQueryByID rebuilds the saved query. When the saved query
# selected output concepts it comes back as a Query; otherwise a
# bare clause / clause-group. All of these run as-is.
loaded_query = session.loadQueryByID("ea5f32d2-2ca8-4820-a96a-41239af75d38")
session.runQuery(loaded_query)
```
