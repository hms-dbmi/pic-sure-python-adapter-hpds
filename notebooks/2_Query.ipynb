```python
import picsure
```

```python
token_file = "token.txt"

with open(token_file, "r") as f:
    my_token = f.read()
```

```python
authorized_session = picsure.connect(
    picsure.Platform.BDC_AUTHORIZED,
    my_token,
    dev_mode=True
)
```

```python
all_facets = authorized_session.showAllFacets()
all_facets
```

```python
tutorial_biolincc_framingham_facet = authorized_session.facets()
tutorial_biolincc_framingham_facet.add('dataset_id', 'tutorial-biolincc_framingham')
```

```python
tutorial_search_results = authorized_session.searchDictionary("Current cigarette smoking at exam", facets=tutorial_biolincc_framingham_facet)
tutorial_search_results['values']
```

```python
cursmoke_var = tutorial_search_results[tutorial_search_results['name'] == 'CURSMOKE']
cursmoke_var
```

```python
raw_values = cursmoke_var['values'].iloc[0]
raw_values
```

```python
cursmoke = picsure.buildClause(
    cursmoke_var['conceptPath'].iloc[0],
    type=picsure.PhenotypicFilterType.FILTER,
    categories=cursmoke_var['values'].iloc[0]
)
```

```python
# Assemble a Query from the filter tree using the new buildQuery() interface.
full_query = picsure.buildQuery(phenotypicFilter=cursmoke)

count_result = authorized_session.runQuery(full_query, type=picsure.QueryType.COUNT)
```

```python
count_result.value
```

```python
# includeConcepts adds output columns without changing who matches —
# this is the replacement for the old SELECT clause. Pull participant
# rows for the cohort, returning the smoking concept as a column.
participant_query = picsure.buildQuery(
    phenotypicFilter=cursmoke,
    includeConcepts=[cursmoke_var["conceptPath"].iloc[0]],
)

participants = authorized_session.runQuery(
    participant_query, type=picsure.QueryType.PARTICIPANT
)
participants.head()
```
