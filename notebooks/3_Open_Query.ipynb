```python
import picsure
```

```python
open_session = picsure.connect(
    picsure.Platform.BDC_OPEN,
    dev_mode=True
)
```

```python
study_facet = open_session.facets()
study_facet.add("dataset_id", "phs004002")
```

```python
study_consents_search_result = open_session.searchDictionary("CONSENT", facets=study_facet)
```

```python
study_consents_search_result['conceptPath'].iloc[0]
```

```python
study_consents_search_result['values'].iloc[0]
```

```python
consentClause = picsure.createSubQuery(
    study_consents_search_result['conceptPath'].iloc[0],
    type=picsure.ClauseType.REQUIRE,
)

consentClause
```

```python
consent_query_result = open_session.runQuery(consentClause, type=picsure.QueryType.CROSS_COUNT)
consent_query_result
```
